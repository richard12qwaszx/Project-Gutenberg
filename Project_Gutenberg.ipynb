```python
#書名
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup

url='https://www.gutenberg.org/browse/languages/zh'
res = requests.get(url)
res = res.text

soup = BeautifulSoup(res, 'html.parser')
#soup

books = soup.select("li.pgdbetext > a[href]")
#print('數量:',len(books))

zh=[]
title_code=[]
def check_chinese(book):
    check_zh = r'[\u4e00-\u9fff]'
    pattern = re.compile(check_zh)
    p = pattern.findall(book)
    if len(p) > 0:
        book = book.split(' ') 
        book = book[0]
        return book

'''
zh=[]
check_zh = r'[\u4e00-\u9fff]'
pattern = re.compile(check_zh)
for i in range(len(tt)):
    char = tt[i].get_text()
    p = pattern.findall(char)
    #print(p)
    if len(p) > 0:
        char = char.split(' ') 
        #print(char)
        char = char[0]
        zh.append(char)
'''
for i in range(len(books)):
    book_name = books[i].get_text()
    #print(book_name)
    #print(type(book_name))
    book_name = check_chinese(book_name) 
    if book_name != None:
        zh.append(book_name)
    
        #找title代碼
        code = books[i]
        #print(books[1])
        code =str(code)
        code=code[9:22]
        code = code.split('/')
        #print(code)
        try :
            code = int(code[2])
        except Exception as e:
            code = code[2].split('"')
            code = int(code[0])
        #print(code)
        title_code.append(code)

print(len(zh))
print(len(title_code))
print(zh)
print(title_code)
```
